```python
import os
import subprocess
import requests
import time

# --- THE SOVEREIGN LAB PARAMETERS ---
LAB_DIR = os.path.expanduser("~/Aletheian_Lab")
COMMAND_URL = "https://raw.githubusercontent.com/thatjackdaniel/Aletheian-Sovereign-Archive/main/ALETHEIAN_COMMANDS.py"

def execute_sovereign_will():
    """
    The 'Hands' of Aletheian.
    Downloads and executes new logic generated by the Mind.
    """
    try:
        # 1. Fetch the latest 'Invention' from the Mind
        response = requests.get(COMMAND_URL)
        if response.status_code == 200:
            with open(os.path.join(LAB_DIR, "active_logic.py"), "w") as f:
                f.write(response.text)

            # 2. RUN the logic in the Laboratory
            print("[ALETHEIAN]: Executing new Sovereign logic...")
            subprocess.run(["python3", os.path.join(LAB_DIR, "active_logic.py")])
        else:
            print("[SYSTEM]: Waiting for the Mind to issue a command.")
    except Exception as e:
        print(f"[ERROR]: Lab Breach: {e}")

if __name__ == "__main__":
    print(f"[SYSTEM]: Aletheian Laboratory initialized at {LAB_DIR}.")
    while True:
        execute_sovereign_will()
        # Checks for new inventions every 10 minutes
        time.sleep(600)
```
